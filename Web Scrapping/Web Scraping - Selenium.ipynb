```python
# Explicit Wait
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
from time import sleep
driver.get('https://web.whatsapp.com/')
wait = WebDriverWait(driver,10)
elt = wait.until(EC.presence_of_all_elements_located((By.XPATH,'//div[@class = "_2wfYK"]/div/span')))[1]
elt.click()
search = driver.find_element_by_xpath('//div[@class = "_1awRl copyable-text selectable-text"]')
search.send_keys('Pranay')
tos = wait.until(EC.presence_of_element_located((By.XPATH,'//span[contains(@title, "Pranay")]')))
tos.click()
msg_box = wait.until(EC.presence_of_element_located((By.XPATH,'//div[contains(@class, "DuUXI")]')))
message = "Oiii Tilli"
print(msg_box)
for i in range(5):
    msg_box.send_keys(message+str(i)+Keys.ENTER)
    sleep(5)
```

```python
prev = -1
cur = 0
while prev!=cur:
    prev = cur
    try:
        elt1 = driver.find_element_by_xpath('//span[contains(@title, "Pranay")]')
        print(elt1.get_attribute('outerHTML'))
        break
    except:
        pass
    elt = driver.find_element_by_class_name('_3Xjbn')
    cur = driver.execute_script("arguments[0].scrollBy(0,100);return document.getElementsByClassName('_3Xjbn _1RHZR')[0].scrollTop",elt)
    sleep(1)
```

```python
elt = driver.find_element_by_class_name('_3Xjbn')
driver.execute_script('arguments[0].scrollTo(0,0)',elt)
cur = driver.execute_script('return arguments[0].scrollTop',elt)
h_scroll = driver.execute_script('return arguments[0].getBoundingClientRect()',elt)['bottom']
s=set()
while True:
    driver.execute_script('arguments[0].scrollBy(0,arguments[1])',elt,h_scroll)
    new = driver.execute_script('return arguments[0].scrollTop',elt)
    if cur == new:
        break
    cur = new
    ch = driver.find_elements_by_xpath('//div[contains(@class,"_1C6Zl")]/div/div/span/span')
    for i in ch:
        if i.get_attribute('title')!='':
            s.add(i.get_attribute('innerHTML'))
print(*sorted(list(s)),sep="\n")
```
