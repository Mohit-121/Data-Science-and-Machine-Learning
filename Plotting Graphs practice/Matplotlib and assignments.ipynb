```python
# Year vs Job opening
df = pd.read_csv('amazon_jobs_dataset.csv')
df1 = df['Posting_date']
df2 = df.Posting_date.str.split(expand=True)[2]
#print(df2.head())
year = df2.unique()
df3 = df.groupby([df2])
df4 = df3.Posting_date.count()
year,count=[],[]
for i in df4.index:
    print(i,df4[i])
    year.append(i)
    count.append(df4[i])
plt.plot(year,count,color="orange",marker="o")
plt.xlabel("Year")
plt.ylabel("No of Job Postings")
plt.title("Year Vs No of Job Postings(Amazon Job Database)")
plt.show()
```

```python
# Month vs Job opening
df = pd.read_csv('amazon_jobs_dataset.csv')
df1 = df['Posting_date']
df2 = df.Posting_date.str.split(expand=True)[0]
df3 = df.groupby([df2])
df4 = df3.Posting_date.count()
month,count=[],[]
for i in df4.index:
    print(i,df4[i])
    month.append(i)
    count.append(df4[i])
plt.bar(month,count,color="green")
plt.xlabel("Month")
plt.ylabel("No of Job Postings")
plt.xticks(rotation=90)
plt.title("Month Vs No of Job Postings(Amazon Job Database)")
plt.show()
```

```python
import matplotlib.pyplot as plt #importing the libraries
import numpy as np
import pandas as pd
df = pd.read_csv('amazon_jobs_dataset.csv')
loc = df['location']
loc = loc.str.split(', ',expand=True)
loc = loc[loc[0]=='IN']
city = loc[2]
city_group = city.value_counts()
# print(type(city_group))
cities = city_group.index
jobs = [city_group[i] for i in cities]
tot = sum(jobs)
for i,j in zip(cities,jobs):
    print(i,round(j*100/tot,2))
explode = [0 for i in range(len(cities))]
explode[0]=0.1
explode[4]=0.1
plt.pie(jobs,labels=cities,explode=explode,startangle=100)
plt.title("Indian Cities Vs Jobs")
plt.show()
```

```python
# Years vs Opening in Java
import matplotlib.pyplot as plt #importing the libraries
import numpy as np
import pandas as pd
df = pd.read_csv('amazon_jobs_dataset.csv')
df = df[df['BASIC QUALIFICATIONS'].notnull()]   #removing null values from basic qualifications
df1 = df[((df['BASIC QUALIFICATIONS'].str.contains("Java")) |
        (df['BASIC QUALIFICATIONS'].str.contains("java")))]  #filtering out java values
df2 = df1['Posting_date'].str.split(expand=True)[2]   #finding job profile with respect to year
opening = df2.value_counts().sort_index()    #finding the count with respect to each year and sort based on year
year,count=[],[]
for i in opening.index:
    print(i,opening[i])
    year.append(i)
    count.append(opening[i])
colors = [i for i in range(len(year))]
plt.scatter(year,count,c=colors)
plt.xlabel("Year")
plt.ylabel("No of openings related to java")
plt.title("Years vs Opening in Java")
plt.show()
```
