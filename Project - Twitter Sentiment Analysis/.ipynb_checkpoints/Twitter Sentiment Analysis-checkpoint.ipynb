## Twitter Sentiment Analysis
Problem Statement: 
1. Perform Sentiment Analysis using Natural Language Processing.
2. Apply different classifiers present in both NLTK Library and SKlearn Library
3. Compare Results of all the classifiers and choose the best suited for the dataset defined

DataSet used: Twitter US Airline Sentiment Analysis - Obtained from Kaggle

DataSet detail: It consists of tweets related to various US Airline in year 2015, and sentiment related to that particular tweet. Size of the dataset is 14,640 rows and 12 columns

Steps:
1. Intially the data obtained is cleaned
2. A dictionary of words with it's corresponding frequency is created using training data
3. The dictionary so created is used to find out most appropriate K words to be used as the filter(We are using 3500 as K)
4. Using the corresponding features we create input array of x_train and x_test for sklearn classifiers and feature dictionary for nltk based classifiers
5. Using various classifiers we predict the accuracy of the model
6. Then we compare each of those results and choose the best out of all

Result:
1. Accuracy using NLTK Naive Bayes with 3500 feature size = 79%
2. Accuracy using Support Vector Machines(SVM) with 3500 feature size = 80%
3. Accuracy using Random Forest with 3500 feature size = 76%
4. Accuracy using Multinomial Naive Bayes with 3500 feature size = 79%
5. Accuracy using K-Nearest Neighbor with 3500 feature size = 74%
6. Accuracy using Logistic Regression using 3500 feature size = 81%

Remarks: Accuracy of 78-80% across testing data is normally assumed to be great in any NLP and Text Classification project, furthermore the increase in number of features and more complicated preprocessing could lead to further increase in accuracy.

```python
# importing the libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# !pip install emot
from emot.emo_unicode import UNICODE_EMO, EMOTICONS

import random
import string
import time
import re

# Run the following 2 lines to download nltk
# import nltk
# nltk.download()
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords, wordnet
from nltk import FreqDist, NaiveBayesClassifier, classify

from sklearn import set_config
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
set_config(print_changed_only = False)
```

```python
# Loading the dataset from the csv file
df = pd.read_csv('Tweets.csv')
```

```python
# Top 5 enteries in the dataset
df.head()
```

```python
# Bottom 5 enteries in the dataset
df.tail()
```

```python
# Getting the relevant fields from the dataset i.e. the text and the airline_sentiment
dataset = df[['text','airline_sentiment']]
```

```python
# renaming the column to text to tweet_text and checking the dataset
dataset = dataset.rename(columns = {'text':'tweet_text'})
dataset.head()
```

```python
# Function for converting emojis into word (eg: 😂😒, etc.. into words)
def convert_emojis(text):
    for emot in UNICODE_EMO:
        text = text.replace(emot, "_".join(UNICODE_EMO[emot].replace(",","").replace(":","").split()))
    return text
```

```python
# Function for converting emoticons into word (eg: :), :>, etc.. into words)
def convert_emoticons(text):
    for emot in EMOTICONS:
        text = re.sub(u'('+emot+')', "_".join(EMOTICONS[emot].replace(",","").split()), text)
    return text
```

```python
print("Various categories of outputs:")
set(dataset['airline_sentiment'])
```

```python
# Creating a stop words set to remove the most occuring and most common stop words
stops = set(stopwords.words('english'))
punctuation = set(string.punctuation)
stops.update(punctuation)
stops
```

```python
def clean_tweet(tweet):
    '''
        argument => Tweet in form of a string
        return the clean_word after removing the stop words, emojis and converting to lower_case
    '''
    txt_tweet = convert_emojis(tweet)
    txt_tweet = convert_emojis(txt_tweet)
    txt_tweet = " ".join([w for w in txt_tweet.split() if not w.startswith('@')])
    words = word_tokenize(txt_tweet)
    output_words = []
    for w in words:
        if w.lower() not in stops:
            output_words.append(w.lower())
    return output_words
```

```python
print("Preprocessing and cleaning the dataset")
percentage,prev, cur = 0, 0, 0
print("0% Completed")
start_time = time.time()

documents = []
for idx in range(len(dataset)):
    # cleaning the dataset
    tweet, sentiment = dataset['tweet_text'][idx], dataset['airline_sentiment'][idx]
    documents.append((clean_tweet(tweet),sentiment))
    
    # Calculating the percentage of the dataset cleaned
    cur += 1
    percentage = cur*100//len(dataset)
    if(prev!=percentage):
        print(str(percentage)+"% Completed")
    prev = percentage

random.seed(1234)
random.shuffle(documents)

end_time = time.time()
print("Time taken to clean the dataset:",(end_time-start_time),"seconds")
len(documents)
```

```python
# Just to see how a particular document tuple looks like tuple([list of words], category)
documents[3]
```

```python
# Defining length of train and test splits
train_len = int(0.8 * len(documents))
test_len = len(documents) - train_len
train_len,test_len
```

```python
# Splitting the data where 80% goes to train and 20% for test
train_doc = documents[:train_len]
test_doc = documents[train_len:]
train_doc[0]
```

```python
test_doc[0]
```

```python
# List of all possible words in the documents
all_words = []
for doc in train_doc:
    all_words += doc[0]
len(all_words)
```

```python
# Frequency distribution of words with {'word':frequency} dictionary for each word
freq_dist = FreqDist(all_words)
```

```python
word_freq = dict(freq_dist)
sort_dict = sorted(word_freq.items(), key = lambda x: -x[1])
print('Top 30 words in the Dictionary with respect to frequency')
print(*sort_dict[:30],sep = '\n')
freq = list(word_freq.values())
freq.sort(reverse = True)
```

```python
# Seeing different graph to figure out best k words
range_xl = [0, 0, 1000, 5000]
range_xh = [len(freq), 100, 5000, 10000]
for i in range(len(range_xl)):
    plt.figure(figsize = (20,7))
    plt.grid()
    plt.xlabel('Nth word')
    plt.ylabel('Frequency')
    plt.title('Frequency distribution graph')
    xl, xh =  range_xl[i], range_xh[i]
    plt.plot(range(len(freq))[xl:xh],freq[xl:xh])
    plt.show()
```

```python
# Using the above graph, we decide to choose top 3500 words as our features
common = freq_dist.most_common(3500)
features = [i[0] for i in common]
```

```python
features
```

```python
def get_feature_dict(words):
    '''
        returns a dict indicating {'feature':True/False}
    '''
    current_features = {}
    words_set = set(words)
    for w in features:
        current_features[w] = w in words_set
    return current_features
```

```python
train_data = [(get_feature_dict(doc),cat) for doc,cat in train_doc]
test_data = [(get_feature_dict(doc),cat) for doc,cat in test_doc]
```

```python
train_data[0]
```

```python
model_name = []
model_accuracy = []
```

## Model - 1 NLTK Naive Bayes Classifier

```python
classifier = NaiveBayesClassifier.train(train_data)
```

```python
accuracy = classify.accuracy(classifier, test_data)
accuracy
```

```python
model_name.append('NLTK Naive Bayes')
model_accuracy.append(accuracy)
```

```python
y_pred = classifier.classify_many(np.array([test_data[i][0] for i in range(len(test_data))]))
y_test = [test_data[i][1] for i in range(len(test_data))]
```

```python
classifier.show_most_informative_features(15)
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

## Converting the data into a format of X - 2D matrix and Y - 1D matrix to use it with SKlearn datasets

```python
# extracting the X and Y values
categories = dataset['airline_sentiment']
text_doc = dataset['tweet_text']
```

```python
x_train, x_test, y_train, y_test = train_test_split(text_doc, categories, random_state = 0)
```

```python
count_vec = TfidfVectorizer(max_features=3500, ngram_range=(1,2), max_df = 0.8, min_df = 3, 
                            use_idf=1,smooth_idf=1, sublinear_tf=1)
x_train_features = count_vec.fit_transform(x_train)
x_train_features.todense()
```

```python
count_vec.get_feature_names()
```

```python
x_test_features = count_vec.transform(x_test)
x_test_features.todense().shape
```

## Model - 2 Support Vector Machines(SVM)

```python
svc = SVC(kernel = 'rbf', probability = True)
svc.fit(x_train_features, y_train)
```

```python
accuracy = svc.score(x_test_features, y_test)
accuracy
```

```python
model_name.append('Support Vector Machines')
model_accuracy.append(accuracy)
```

```python
y_pred = svc.predict(x_test_features)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

## Model - 3 Random Forest Classifier

```python
depths = [1, 2, 4, 8, 16, 32, 64, 128, 180]
test_accuracy, train_accuracy = [], []
for depth in depths:
    rfc = RandomForestClassifier(max_depth = depth, random_state = 0, n_jobs = -1)
    rfc.fit(x_train_features, y_train)
    train_accuracy.append(rfc.score(x_train_features, y_train))
    test_accuracy.append(rfc.score(x_test_features, y_test))
```

```python
plt.figure(figsize = (12,8))
plt.grid()
plt.plot(depths,train_accuracy,c = 'b')
plt.plot(depths,test_accuracy, c= 'r')
plt.xlabel('Depth')
plt.ylabel('Accuracy')
plt.legend(['Train accuracy', 'Test accuracy'])
plt.title('Random Forest - Depth Vs Accuracy plot')
plt.show()
```

```python
estimators = [1, 2, 4, 8, 16, 32, 64, 128]
test_accuracy, train_accuracy = [], []
for estimator in estimators:
    rfc = RandomForestClassifier(n_estimators = estimator, random_state = 0, n_jobs = -1)
    rfc.fit(x_train_features, y_train)
    train_accuracy.append(rfc.score(x_train_features, y_train))
    test_accuracy.append(rfc.score(x_test_features, y_test))
```

```python
plt.figure(figsize = (12,8))
plt.grid()
plt.plot(estimators,train_accuracy,c = 'b')
plt.plot(estimators,test_accuracy, c= 'r')
plt.xlabel('No of Estimators')
plt.ylabel('Accuracy')
plt.legend(['Train accuracy', 'Test accuracy'])
plt.title('Random Forest - No of Estimators Vs Accuracy plot')
plt.show()
```

```python
rfc = RandomForestClassifier(n_estimators = 128, max_depth = 128, random_state = 0, n_jobs = -1)
rfc.fit(x_train_features, y_train)
```

```python
accuracy = rfc.score(x_test_features, y_test)
accuracy
```

```python
model_name.append('Random Forest')
model_accuracy.append(accuracy)
```

```python
y_pred = rfc.predict(x_test_features)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

## Checking Accuracy on any given Sample Tweet using the Model - 1 NLTK Naive Bayes Classifier

```python
def check_prediction_naive_bayes(text):
    cleaned_text = clean_tweet(text)
    text_features = get_feature_dict(cleaned_text)
    dist = classifier.prob_classify(text_features)
    dict_list = ['negative','neutral','positive']
    print("Probability using naive_bayes:")
    print(" ".join(dict_list))
    print("{0:.3f}    {1:.3f}   {2:.3f}" .format(dist.prob('negative'),dist.prob('neutral'),dist.prob('positive')))
    return classifier.classify(text_features)
```

```python
def check_prediction_svm(text):
    text_features = count_vec.transform([text])
    dict_list = ['negative','neutral','positive']
    probs = svc.predict_proba(text_features)[0]
    print("Probability using Support Vector Machine(SVM):")
    print(" ".join(dict_list))
    print("{0:.3f}    {1:.3f}   {2:.3f}" .format(probs[0],probs[1],probs[2]))
    return svc.predict(text_features)[0]
```

```python
def check_prediction_lr(text):
    text_features = count_vec.transform([text])
    dict_list = ['negative','neutral','positive']
    probs = lr.predict_proba(text_features)[0]
    print("Probability using LogisticRegression:")
    print(" ".join(dict_list))
    print("{0:.3f}    {1:.3f}   {2:.3f}" .format(probs[0],probs[1],probs[2]))
    return svc.predict(text_features)[0]
```

```python
text = "not good, need a bit of improvement"
print(check_prediction_naive_bayes(text))
print('------------------------------------')
print(check_prediction_svm(text))
print('------------------------------------')
print(check_prediction_lr(text))
```
