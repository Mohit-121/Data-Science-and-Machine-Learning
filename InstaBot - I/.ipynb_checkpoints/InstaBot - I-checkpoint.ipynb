2) Type for “food” in search bar and print all the names of the Instagram Handles that are displayed in list after typing “food”

```python
# 5) Finding and opening dilsefoodie instagram handle
search = driver.find_element_by_class_name('XTCLo')
# clear the search bar and type food and press enter
search.clear()
search.send_keys('dilsefoodie'+Keys.ENTER)
elt = wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class = "fuqBx"]//span[@class = "Ap253"]')))
elt.click()
```

```python
# 5.1) Liking the top 30 posts of the ‘dilsefoodie'. 
#      Print message if you have already liked it.
# close = driver.find_element_by_class_name('yiMZG')
# close.click()
posts = driver.find_elements_by_class_name('_9AhH0')
c=0
for post in posts:
    post.click()
    like = wait.until(EC.presence_of_element_located((By.XPATH,'//span[@class = "fr66n"]/button/div/span/*[name()="svg"]')))
    lk_ulk = like.get_attribute('aria-label')
    if lk_ulk == 'Unlike':
        print('Post Already Liked')
    else:
        like.click()
    close = driver.find_element_by_class_name('yiMZG')
    close.click()
    c+=1
    if(c==30):
        break
# print(c)  Testing for count of posts
```

```python
# 5.2) Unliking the top 30 posts of the ‘dilsefoodie’. 
#      Print message if you have already unliked it.
posts = driver.find_elements_by_class_name('_9AhH0')
c=0
for post in posts:
    post.click()
    like = wait.until(EC.presence_of_element_located((By.XPATH,'//span[@class = "fr66n"]/button/div/span/*[name()="svg"]')))
    lk_ulk = like.get_attribute('aria-label')
    if lk_ulk == 'Unlike':
        like.click()
    else:
        print('Post Already Unliked')
    close = driver.find_element_by_class_name('yiMZG')
    close.click()
    c+=1
    if(c==30):
        break
```

```python
# 6.1) Extract the usernames of the first 500 followers of ‘foodtalkindia’
# search and find foodtalkindia
driver.get('https://www.instagram.com/')
search = driver.find_element_by_class_name('XTCLo')
#clear the search bar and type food and press enter
search.clear()
search.send_keys('foodtalkindia'+Keys.ENTER)
elt = wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class = "fuqBx"]//span[@class = "Ap253"]')))
elt.click()

# finding the followers
fol_list = wait.until(EC.presence_of_element_located((By.PARTIAL_LINK_TEXT,'followers')))
fol_list.click()
c=0      # count of followers
f_span = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'isgrP')))
while c<500:
    elts = wait.until(EC.presence_of_all_elements_located((By.XPATH,'//a[contains(@class , "FPmhX")]')))
    h = driver.execute_script("return arguments[0].scrollHeight",f_span)
    c = len(elts)
    driver.execute_script("arguments[0].scrollBy(0,arguments[1])",f_span,h)
    sleep(2)
c=0
# printing the names of first 500 followers of foodtalkindia
fti_fol = []
for i in elts:
    c+=1
    fol = i.get_attribute('innerHTML')
    print(c,fol)
    fti_fol.append(fol)
    if c==500:
        break
```

```python
# # 6.1) Extract the usernames of the first 500 followers of ‘sodelhi’
# search and find foodtalkindia
driver.get('https://www.instagram.com/')
search = driver.find_element_by_class_name('XTCLo')
#clear the search bar and type food and press enter
search.clear()
search.send_keys('sodelhi'+Keys.ENTER)
elt = wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class = "fuqBx"]//span[@class = "Ap253"]')))
elt.click()

# finding the followers
fol_list = wait.until(EC.presence_of_element_located((By.PARTIAL_LINK_TEXT,'followers')))
fol_list.click()
c=0      # count of followers
f_span = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'isgrP')))
while c<500:
    elts = wait.until(EC.presence_of_all_elements_located((By.XPATH,'//a[contains(@class , "FPmhX")]')))
    h = driver.execute_script("return arguments[0].scrollHeight",f_span)
    c = len(elts)
    driver.execute_script("arguments[0].scrollBy(0,arguments[1])",f_span,h)
    sleep(2)
c=0
# printing the names of first 500 followers of sodelhi
for i in elts:
    c+=1
    print(c,i.get_attribute('innerHTML'))
    if c==500:
        break
```

```python
# finding my followers and following
m_fol = 'https://www.instagram.com/'+suname
driver.get(m_fol)
my_follower = []
my_following = []

# finding the followers
fol_list = wait.until(EC.presence_of_element_located((By.PARTIAL_LINK_TEXT,'followers')))
fol_list.click()
p=0      # count of followers
f_span = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'isgrP')))
while True:
    elts = wait.until(EC.presence_of_all_elements_located((By.XPATH,'//a[contains(@class , "FPmhX")]')))
    h = driver.execute_script("return arguments[0].scrollHeight",f_span)
    c = len(elts)
    if p == c:
        break
    driver.execute_script("arguments[0].scrollBy(0,arguments[1])",f_span,h)
    sleep(2)
    p = c

c=0
# adding followers to my_follower list
for i in elts:
    c+=1
    my_follower.append(i.get_attribute('innerHTML'))
    
driver.find_elements_by_class_name('WaOAr')[1].click()
    
#finding the following
fol_list = wait.until(EC.presence_of_element_located((By.PARTIAL_LINK_TEXT,'following')))
fol_list.click()
p=0      # count of followers
f_span = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'isgrP')))
while True:
    elts = wait.until(EC.presence_of_all_elements_located((By.XPATH,'//a[contains(@class , "FPmhX")]')))
    h = driver.execute_script("return arguments[0].scrollHeight",f_span)
    c = len(elts)
    if p == c:
        break
    driver.execute_script("arguments[0].scrollBy(0,arguments[1])",f_span,h)
    sleep(2)
    p = c


c=0
# adding to my_following list
for i in elts:
    c+=1
    my_following.append(i.get_attribute('innerHTML'))
driver.find_elements_by_class_name('WaOAr')[1].click()
```

```python
# 6.2) Now print all the followers of “foodtalkindia” that you are following 
# but those who don’t follow you.
l1 = []
for i in fti_fol:
    if i in my_following:
        l1.append(i)
l2 = []
for i in l1:
    if i not in my_follower:
        l2.append(i)
print('List of people who are folowers of foodtalkindia that I am following but who are not following me:')
if len(l2)==0:
    print("No match found!")
else:
    print(*l2,sep=" ")
```

```python
# 7) Check the story of ‘coding.ninjas’. Consider the following Scenarios and print error messages accordingly -

    # 7.1) If You have already seen the story.
    # 7.2) Or The user has no story.
    # 7.3) Or View the story if not yet seen.
```

```python
# search and find coding.ninjas
driver.get('https://www.instagram.com/')
search = driver.find_element_by_class_name('XTCLo')
#clear the search bar and type food and press enter
search.clear()
search.send_keys('coding.ninjas'+Keys.ENTER)
elt = wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class = "fuqBx"]//span[@class = "Ap253"]')))
elt.click()
```

```python
try:
    # If user has a story view it
    img = wait.until(EC.presence_of_element_located((By.XPATH,'//div[contains(@class,"h5uC0")]')))
    driver.maximize_window()
    canvas = img.find_element_by_tag_name('canvas')
    att = canvas.get_attribute('style').split('; ')
    d = {}
    for i in att:
        k,v = i.split(":")
        d[k.strip()]=v.strip()
    # If haven't viewed the story
    if d['height'] == '168px;':
        print('You have not viewed the story, opening the story!')
        img.click()
    # If already viewed the story
    else:
        print('You have already viewed this users story!')
except:
    print('User has no story!')   # In this case user has no story
```
