## Text Classification
Problem Statement: 
1. Perform Test Classification using Multinomial Naive Bayes(already implemented in sklearn)
2. Implement Naive Bayes on your own from scratch for text classification. 
3. Compare Results of your implementation of Naive Bayes with one in Sklearn.

Steps:
1. Intially the data obtained is cleaned
2. A dictionary of words with it's corresponding frequency is created using X_train data
3. The dictionary so created is used to find out most appropriate k words to be used as the filter
4. Using the corresponding features we create input array of X_train and X_test
5. Using inbuilt Multinomial Naive Bayes we predict the values
6. Then creating our own implementation of the Naive Bayes and checking the predictions of those values and comparing it with inbuilt class predictions

Result:
1. Accuracy using inbuilt Multinomial Naive Bayes with 10,000 feature size = 0.81
2. Accuracy using my implemention of Naive Bayes with 10,000 feature size = 0.85

Remarks: Accuracy of 0.85 across testing data is normally assumed to be great, furthermore the increase in number of features and more complicated preprocessing could lead to further increase in accuracy.

P.S -: It takes around 10-15 minutes on any normal system to run this project completely, hence, request the TA to be a bit more patient with the process.

```python
# Importing all the neccessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import string
from nltk.corpus import stopwords, wordnet
from nltk import pos_tag, word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.stem.porter import PorterStemmer
from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix,classification_report
from datetime import datetime
```

```python
def clean_text(file):
    '''
        The cleaned version of file wherein the the punctuations and stop words are
        removed and a standard casing is used, additionally we try to stem the word
        and then return those words for the given file
    '''
    stop_words = set(stopwords.words('english'))
    punctuations = set(string.punctuation)
    stop_words.update(punctuations)
    porter = PorterStemmer()
    words = word_tokenize(file)
    words = [word.lower() for word in words if len(word)>2 and word.isalpha() and word.lower() not in stop_words]
    words = [porter.stem(word) for word in words if word not in stop_words]
    return words
```

```python
# Fetch the data and perform preprocessing on the data
newsgroup = fetch_20newsgroups()
X = newsgroup.data
Y = newsgroup.target

start_time = datetime.now()

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state = 1)
X_train_words, X_test_words = [],[]        # will contain the cleaned words of the X_train and X_test respectively
word_freq = {}                             # Denotes the final freq dictionary of the words in all the documents
percentage,prev, cur = 0, 0, 0
print("Training Data Preprocessing")
print("0% Completed")
for file in X_train:
    cleaned = clean_text(file)
    X_train_words.append(cleaned)
    for word in cleaned:
        word_freq[word] = word_freq.get(word,0)+1
    cur += 1
    percentage = cur*100//len(X_train)
    if(prev!=percentage):
        print(str(percentage)+"% Completed")
    prev = percentage

print("Testing Data Preprocessing")
print("0% Completed")
percentage,prev, cur = 0, 0, 0
for file in X_test:
    cleaned = clean_text(file)
    X_test_words.append(cleaned)
    cur += 1
    percentage = cur*100//len(X_test)
    if(prev!=percentage):
        print(str(percentage)+"% Completed")
    prev = percentage

end_time = datetime.now()
dif = end_time - start_time
print('Time take to perform preprocessing is:', (dif.seconds//60), 'minutes', (dif.seconds%60), 'seconds')
    
print('Length of words dictionary is:',len(word_freq))
```

```python
sort_dict = sorted(word_freq.items(), key = lambda x: -x[1])
print('Top 30 words in the Dictionary with respect to frequency')
print(*sort_dict[:30],sep = '\n')
freq = list(word_freq.values())
freq.sort(reverse = True)
```

## Frequency distribution of words is made and is used to come to the best K words which could be used as features

```python
# Seeing different graph to figure out best k words
range_xl = [0, 0, 1000, 5000]
range_xh = [len(freq), 100, 5000, 10000]
for i in range(len(range_xl)):
    plt.figure(figsize = (20,7))
    plt.grid()
    plt.xlabel('Nth word')
    plt.ylabel('Frequency')
    plt.title('Frequency distribution graph')
    xl, xh =  range_xl[i], range_xh[i]
    plt.plot(range(len(freq))[xl:xh],freq[xl:xh])
    plt.show()
```

```python
# using the above graph the range which looked most prominent was 30, 10030 wherein we take 10000 features(words)
l, h = 30, 10030
word_list = [i for i,j in sort_dict[l:h]]
feature_dict = {word_list[i]:i for i in range(len(word_list))}
print('Length of total features selected:',len(feature_dict))
```

```python
def create_feature_array(X_words, feature_dict):
    '''
        Using the feature_dict we create the modified X array where X[i][j]
        represents for ith row and jth feature, frequency of jth feature(word)
        in ith document.
        return this modified X array
    '''
    X_modified = []
    for words in X_words:
        X1 = np.zeros(len(feature_dict.keys()),dtype = int)
        for word in words:
            if(word in feature_dict.keys()):
                X1[feature_dict[word]] = feature_dict.get(word, 0) + 1
        X_modified.append(X1)
    
    return np.array(X_modified)  
```

```python
# Performing the modifications to get X_train to the desired input format
X_train = create_feature_array(X_train_words,feature_dict)
print('Shape of X training data',X_train.shape)
print('Shape of Y training data',Y_train.shape)
```

```python
# Performing the modifications to get X_test to the desired input format
X_test = create_feature_array(X_test_words,feature_dict)
print('Shape of X testing data',X_test.shape)
print('Shape of Y testing data',Y_test.shape)
```

## Using inbuilt Naive Bayes Classifier and checking it's predictions

```python
clf = MultinomialNB()
clf.fit(X_train,Y_train)
Y_pred = clf.predict(X_test)
print(classification_report(Y_test,Y_pred))
print(confusion_matrix(Y_test,Y_pred))
```

```python
def fit(X_train, Y_train):
    '''
        returns the dictionary where 
        1st level keys => Classes
        2nd level keys => The features
        and 2nd level values => The word_count of a ith class having jth feature
    '''
    result = {}
    classes = set(Y_train)
    result["total_data"] = len(Y_train)
    for current_class in classes:
        current_class_rows = (Y_train == current_class)
        X_train_current = X_train[current_class_rows]
        Y_train_current = Y_train[current_class_rows]
        num_features = X_train.shape[1]
        result[current_class] = {}
        result[current_class]["total_data"] = len(Y_train_current)
        word_count = 0
        for j in range(num_features):
            result[current_class][j] = X_train_current[:,j].sum()
            word_count += result[current_class][j]
        result[current_class]["word_count"] = word_count
    
    return result   
```

```python
def probability(dictionary, x, current_class):
    '''
        returns the probabilty of X given features x where y is the current class
    '''
    output = np.log(dictionary[current_class]['total_data']) - np.log(dictionary['total_data'])
    num_features = len(dictionary[current_class].keys()) - 2
    for j in range(num_features):
        if(x[j]!=0):
            count_current_class_wj = dictionary[current_class][j] + 1
            count_current_class_words = dictionary[current_class]['word_count'] + num_features
            current_wj_probability = np.log(count_current_class_wj) - np.log(count_current_class_words)
            output += current_wj_probability
    return output
```

```python
def predictSinglePoint(dictionary, x):
    '''
        returns the best class for a particular data point
    '''
    classes = dictionary.keys()
    best_p, best_class = -1001,-1
    first_run = True
    for current_class in classes:
        if(current_class == 'total_data'):
            continue
        p_current_class = probability(dictionary, x, current_class)
        if(first_run or p_current_class > best_p):
            best_p = p_current_class
            best_class = current_class
        first_run = False
    return best_class
```

```python
def predict(dictionary, X_test):
    '''
         returns the prediction array for all the testing points
    '''
    y_pred = []
    print("0% Completed")
    percentage,prev, cur = 0, 0, 0
    for x in X_test:
        x_class = predictSinglePoint(dictionary, x)
        y_pred.append(x_class)
        cur += 1
        percentage = cur*100//len(X_test)
        if(prev!=percentage):
            print(str(percentage)+"% Completed")
        prev = percentage
    return y_pred
```

```python
# Fit function which returns the dictionary which could be used while prediction
dictionary = fit(X_train,Y_train)
```

```python
# Predict function which predicts the values for the given X data
start_time = datetime.now()
Y_pred = predict(dictionary,X_test)
end_time = datetime.now()
dif = end_time - start_time
print('Time taken to predict the values for the given x_test is: ', (dif.seconds//60),'minutes and', (dif.seconds%60),'seconds')
```

```python
print(classification_report(Y_test,Y_pred))
print(confusion_matrix(Y_test,Y_pred))
```
