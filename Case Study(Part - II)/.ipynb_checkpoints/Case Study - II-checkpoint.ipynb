```python
# importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Getting the data in a dataframe
df = pd.read_csv('startup_funding.csv')
```

Q1) Your Friend has developed the Product and he wants to establish the product startup and he is searching for a perfect       location where getting the investment has a high chance. But due to its financial restriction, he can choose only between 
three locations -  Bangalore, Mumbai, and NCR. As a friend, you want to help your friend deciding the location. NCR include 
Gurgaon, Noida and New Delhi. Find the location where the most number of funding is done. That means, find the location  where startups has received funding maximum number of times. Plot the bar graph between location and number of funding. 
Take city name "Delhi" as "New Delhi". Check the case-sensitiveness of cities also. That means, at some place instead of 
"Bangalore", "bangalore" is given. Take city name as "Bangalore". For few startups multiple locations are given, one Indian 
and one Foreign. Consider the startup if any one of the city lies in given locations.

```python
# City vs No of Number of Startups
city = df['CityLocation'].str.title().replace('Delhi','New Delhi')  # handles case sensitivenes and the case faults in dataset
city = city.str.split('/',expand=True)   # finds and splits if there are multiple cities present
city = city[((city[0].isnull()==False) | (city[1].isnull()==False))] # check if there is at least one city in the row
city_count = {}
for i in city.values:
    c1,c2=i[0],i[1]
    if(c1!=None):
        c1=c1.strip()
        city_count[c1]=city_count.get(c1,0)+1
    if(c2!=None):
        c2=c2.strip()
        city_count[c2]=city_count.get(c2,0)+1
sorted_city_count = sorted(city_count.items(),key = lambda x: -x[1])[:10]  #get top 10 cities with respect to no of startups
city_list = [sorted_city_count[i][0] for i in range(10)]      # getting the top 10 cities
startup_count = [sorted_city_count[i][1] for i in range(10)]  # getting the no of startups

# Drawing the bar graph to represent the distribution among these top 10 cities
plt.bar(city_list,startup_count)
plt.xlabel("Cities")
plt.ylabel("No of Startups")
plt.title("Cities Vs No of Startups")
plt.xticks(rotation=90)
plt.show()

# Drawing the conclusions(Finding the no of startups)
banglore_startup = city_count['Bangalore']
mumbai_startup = city_count['Mumbai']
ncr_startup = city_count['New Delhi'] + city_count['Gurgaon'] + city_count['Noida']
print("No of Startups in Bangalore","=",banglore_startup,sep="\t")
print("No of Startups in Mumbai","=",mumbai_startup,sep="\t")
print("No of Startups in NCR   ","=",ncr_startup,sep="\t")
```

Q2) Even after trying for so many times, your friend’s startup could not find the investment. So you decided to take this matter in your hand and try to find the list of investors who probably can invest in your friend’s startup. Your list will increase the chance of your friend startup getting some initial investment by contacting these investors. Find the top 5 investors who have invested maximum number of times (consider repeat investments in one company also). In a startup, multiple investors might have invested. So consider each investor for that startup. Ignore undisclosed investors.

```python
# Investor Vs No of Investments
investor_name = df['InvestorsName']
# removing undisclosed investors and null values and split the investors to get all the investors
inv_group = investor_name[((investor_name!='Undisclosed Investors') & (investor_name!='Undisclosed investors'))]
inv_group = inv_group[inv_group.isnull()==False]
inv_group = inv_group.str.strip()
inv_group = inv_group.str.split(',')
inv_count={}      # dictonary to indicate investors and no of their investments
for i in inv_group:
    for j in i:
        val=j.strip()
        inv_count[val]=inv_count.get(val,0)+1
top_investor = sorted(inv_count.items(),key = lambda x: -x[1])[:5]
investor_name = [i[0] for i in top_investor]   # we get top 5 investor from the dict
no_of_investment = [i[1] for i in top_investor]  # we get investment of top 5 investors

# Drawing the bar graph to show the investment by top 5 investors
plt.bar(investor_name,no_of_investment)
plt.xlabel("Investor Name")
plt.ylabel("No of Investments")
plt.title("Investor Vs No of Investments")
plt.xticks(rotation=90)
plt.show()
```

Q3) After re-analysing the dataset you found out that some investors have invested in the same startup at different number of funding rounds. So before finalising the previous list, you want to improvise it by finding the top 5 investors who have invested in different number of startups. This list will be more helpful than your previous list in finding the investment for your friend startup. Find the top 5 investors who have invested maximum number of times in different companies. That means, if one investor has invested multiple times in one startup, count one for that company. There are many errors in startup names. Ignore correcting all, just handle the important ones - Ola, Flipkart, Oyo and Paytm.

```python
df1 = df.copy()   # creating copy of orignal dataset
def handleNames(x):                                    # function to correct the Startup Names
    if x == 'Ola Cabs': return 'Ola'
    if x == 'Olacabs': return 'Ola'
    if x == 'Flipkart.com': return 'Flipkart'
    if x == 'Oyo Rooms': return 'Oyo'
    if x == 'Oyorooms': return 'Oyo'
    if x == 'OYO Rooms': return 'Oyo'
    if x == 'OyoRooms': return 'Oyo'
    if x == 'Paytm Marketplace': return 'Paytm'
    return x

def handleInvestors(x):                                        # function to remove undisclosed Investors
    if x == 'Undisclosed Investors' or x == 'Undisclosed investors':
        return None
    return x

df1['StartupName'] = df1['StartupName'].apply(handleNames) # correcting the startupnames
df1['InvestorsName'] = df1['InvestorsName'].apply(handleInvestors)  # removing the undisclosed investors
df1 = df1[df1['InvestorsName'].isnull()==False]
df1 = df1[df1['StartupName']!='']
inv_group = {}   # inv_group key: investors value: startup name set 
for i in df1.index:
    name,investors=df1['StartupName'][i],df1['InvestorsName'][i].strip()
    investors = investors.split(",")
    for j in investors:
        j=j.strip()
        if j!= '':
            if j not in inv_group:
                s=set()
                inv_group[j]=s
            inv_group[j].add(name)
            
inv_group_len={}
for i in inv_group:
    inv_group_len[i]=len(inv_group[i])
top_investor = sorted(inv_group_len.items(), key = lambda x: -x[1])[:5]

investor_name = [i[0] for i in top_investor]   # we get top 5 investor from the dict
no_of_investment = [i[1] for i in top_investor]  # we get investment of top 5 investors

# Drawing the bar graph to show the investment by top 5 investors in different startups
plt.bar(investor_name,no_of_investment)
plt.xlabel("Investor Name")
plt.ylabel("No of Investments")
plt.title("Investor Vs No of Unique Investments")
plt.xticks(rotation=90)
plt.show()
```

Q4) Even after putting so much effort in finding the probable investors, it didn't turn out to be helpful for your friend. So you went to your investor friend to understand the situation better and your investor friend explained to you about the different Investment Types and their features. This new information will be helpful in finding the right investor. Since your friend startup is at an early stage startup, the best-suited investment type would be - Seed Funding and Crowdfunding. Find the top 5 investors who have invested in a different number of startups and their investment type is Crowdfunding or Seed Funding. Correct spelling of investment types are - "Private Equity", "Seed Funding", "Debt Funding", and "Crowd Funding". Keep an eye for any spelling mistake. You can find this by printing unique values from this column. There are many errors in startup names. Ignore correcting all, just handle the important ones - Ola, Flipkart, Oyo and Paytm.

```python
df1 = df.copy()
def handleNames(x):                         # function to correct the Startup Names
    if x == 'Ola Cabs': return 'Ola'
    if x == 'Olacabs': return 'Ola'
    if x == 'Flipkart.com': return 'Flipkart'
    if x == 'Oyo Rooms': return 'Oyo'
    if x == 'Oyorooms': return 'Oyo'
    if x == 'OYO Rooms': return 'Oyo'
    if x == 'OyoRooms': return 'Oyo'
    if x == 'Paytm Marketplace': return 'Paytm'
    return x

def handleInvestors(x):                                 # function to remove undisclosed Investors
    if x == 'Undisclosed Investors' or x == 'Undisclosed investors':
        return None
    return x

def handleInvestmentType(x):                                # correcting the spellings of investment types
    if x == 'PrivateEquity': return 'Private Equity'
    if x == 'SeedFunding': return 'Seed Funding'
    if x == 'Crowd funding': return 'Crowd Funding'
    return x

df1['StartupName'] = df1['StartupName'].apply(handleNames) # correcting the startupnames
df1['InvestorsName'] = df1['InvestorsName'].apply(handleInvestors)  # removing undisclosed investors
df1['InvestmentType'] = df1['InvestmentType'].apply(handleInvestmentType)  # correcting investment type spellings
df1 = df1[df1['InvestmentType'].isnull()==False]  
df1 = df1[df1['InvestorsName'].isnull()==False]

# finding the crowd funding and seed funding investments and removing startupname as empty string
df1 = df1[((df1['InvestmentType']=='Crowd Funding') | (df1['InvestmentType']=='Seed Funding'))]
df1 = df1[df1['StartupName']!='']
inv_group = {}  # inv_group key: investors value: startup name set 
for i in df1.index:
    name,investors=df1['StartupName'][i],df1['InvestorsName'][i].strip()
    investors = investors.split(',')
    for j in investors:
        j=j.strip()
        if j!= '':
            if j not in inv_group:
                s=set()
                inv_group[j]=s
            inv_group[j].add(name)
            
inv_group_len={}
for i in inv_group:
    inv_group_len[i]=len(inv_group[i])   # len of set gives different unique startups

top_investor = sorted(inv_group_len.items(), key = lambda x: -x[1])[:5]
investor_name = [i[0] for i in top_investor]   # we get top 5 investor from the dict
no_of_investment = [i[1] for i in top_investor]  # we get investment of top 5 investors

# Drawing the bar graph to show the investment by top 5 investors
plt.bar(investor_name,no_of_investment)
plt.xlabel("Investor Name")
plt.ylabel("No of Investments")
plt.title("Investor Vs No of Unique Investments(Crowd and Seed Funding)")
plt.xticks(rotation=90)
plt.show()
```
